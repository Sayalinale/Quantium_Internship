```python
# Example placeholder data
transaction_data = {
    'DATE': pd.date_range(start='2022-01-01', periods=100),
    'TOT_SALES': np.random.randint(100, 1000, size=100),
    'PROD_QTY': np.random.randint(1, 10, size=100)
}
customer_data = {
    'LYLTY_CARD_NBR': np.random.randint(1000, 2000, size=100),
    'LIFESTAGE': np.random.choice(['MIDAGE SINGLES/COUPLES', 'NEW FAMILIES', 'OLDER FAMILIES'], size=100),
    'PREMIUM_CUSTOMER': np.random.choice(['Budget', 'Mainstream', 'Premium'], size=100)
}
transaction_df = pd.DataFrame(transaction_data)
customer_df = pd.DataFrame(customer_data)

# Total sales revenue
total_sales_revenue = transaction_df['TOT_SALES'].sum()

# Average transaction value
average_transaction_value = transaction_df['TOT_SALES'].mean()

# Sales growth over time
sales_growth_over_time = (transaction_df['TOT_SALES'].iloc[-1] - transaction_df['TOT_SALES'].iloc[0]) / transaction_df['TOT_SALES'].iloc[0] * 100

# Customer retention rate (example calculation)
total_customers_start = customer_df['LYLTY_CARD_NBR'].nunique()
total_customers_end = customer_df['LYLTY_CARD_NBR'].nunique()  # Assuming data represents a fixed time period
customer_retention_rate = (total_customers_end / total_customers_start) * 100

# Customer acquisition cost (example calculation)
total_marketing_cost = 10000  # Example total marketing cost
total_new_customers = 50  # Example total new customers acquired
customer_acquisition_cost = total_marketing_cost / total_new_customers

# Customer lifetime value (example calculation)
average_customer_lifetime = 5  # Example average customer lifetime in years
average_revenue_per_customer = total_sales_revenue / total_customers_start  # Example average revenue per customer
customer_lifetime_value = average_revenue_per_customer * average_customer_lifetime

# Print KPIs
print("Key Performance Indicators (KPIs):")
print(f"Total Sales Revenue: {total_sales_revenue:.2f}")
print(f"Average Transaction Value: {average_transaction_value:.2f}")
print(f"Sales Growth Over Time: {sales_growth_over_time:.2f}%")
print(f"Customer Retention Rate: {customer_retention_rate:.2f}%")
print(f"Customer Acquisition Cost: {customer_acquisition_cost:.2f}")
print(f"Customer Lifetime Value: {customer_lifetime_value:.2f}")
```
